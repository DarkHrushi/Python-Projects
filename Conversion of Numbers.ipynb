```python
# Importing Necessary Libraries
import tkinter as tk
from tkinter import messagebox

# Creating a function to convert decimal to binary
def decimal_to_binary(decimal):
    return bin(decimal)[2:]

# Creating a function to convert decimal to octal
def decimal_to_octal(decimal):
    return oct(decimal)[2:]

# Creating a function to convert decimal to hexadecimal
def decimal_to_hexadecimal(decimal):
    return hex(decimal)[2:]

# Creating a function to convert binary to decimal
def binary_to_decimal(binary):
    return int(binary, 2)

# Creating a function to convert binary to octal
def binary_to_octal(binary):
    decimal = binary_to_decimal(binary)
    return decimal_to_octal(decimal)

# Creating a function to convert binary to hexadecimal
def binary_to_hexadecimal(binary):
    decimal = binary_to_decimal(binary)
    return decimal_to_hexadecimal(decimal)

# Creating a function to convert octal to decimal
def octal_to_decimal(octal):
    return int(octal, 8)

# Creating a function to convert octal to binary
def octal_to_binary(octal):
    decimal = octal_to_decimal(octal)
    return decimal_to_binary(decimal)

# Creating a function to convert octal to hexadecimal
def octal_to_hexadecimal(octal):
    decimal = octal_to_decimal(octal)
    return decimal_to_hexadecimal(decimal)

# Creating a function to convert hexadecimal to decimal
def hexadecimal_to_decimal(hexadecimal):
    return int(hexadecimal, 16)

# Creating a function to convert hexadecimal to binary
def hexadecimal_to_binary(hexadecimal):
    decimal = hexadecimal_to_decimal(hexadecimal)
    return decimal_to_binary(decimal)

# Creating a function to convert hexadecimal to octal
def hexadecimal_to_octal(hexadecimal):
    decimal = hexadecimal_to_decimal(hexadecimal)
    return decimal_to_octal(decimal)

def convert():
    input_val = input_entry.get()
    choice = option_var.get()
    output_val = ""

    try:
        if choice == "Decimal":
            decimal = int(input_val)
            output_val += "Binary: " + decimal_to_binary(decimal) + "\n"
            output_val += "Octal: " + decimal_to_octal(decimal) + "\n"
            output_val += "Hexadecimal: " + decimal_to_hexadecimal(decimal)
        elif choice == "Binary":
            output_val += "Decimal: " + str(binary_to_decimal(input_val)) + "\n"
            output_val += "Octal: " + binary_to_octal(input_val) + "\n"
            output_val += "Hexadecimal: " + binary_to_hexadecimal(input_val)
        elif choice == "Octal":
            output_val += "Decimal: " + str(octal_to_decimal(input_val)) + "\n"
            output_val += "Binary: " + octal_to_binary(input_val) + "\n"
            output_val += "Hexadecimal: " + octal_to_hexadecimal(input_val)
        elif choice == "Hexadecimal":
            output_val += "Decimal: " + str(hexadecimal_to_decimal(input_val)) + "\n"
            output_val += "Binary: " + hexadecimal_to_binary(input_val) + "\n"
            output_val += "Octal: " + hexadecimal_to_octal(input_val)
        
        output_text.config(state=tk.NORMAL)
        output_text.delete(1.0, tk.END)
        output_text.insert(tk.END, output_val)
        output_text.config(state=tk.DISABLED)
    except ValueError:
        messagebox.showerror("Error", "Invalid input")

# Creating a main window
root = tk.Tk()
root.title("Number Conversion Tool")

# Input frame
input_frame = tk.Frame(root)
input_frame.pack(pady=10)

input_label = tk.Label(input_frame, text="Enter number:")
input_label.grid(row=0, column=0)

input_entry = tk.Entry(input_frame, width=30)
input_entry.grid(row=0, column=1)

# Option menu for selecting number system
option_var = tk.StringVar()
option_label = tk.Label(input_frame, text="Select number system:")
option_label.grid(row=1, column=0)

option_menu = tk.OptionMenu(input_frame, option_var, "Decimal", "Binary", "Octal", "Hexadecimal")
option_menu.grid(row=1, column=1)
option_var.set("Decimal")

# Convert button
convert_button = tk.Button(root, text="Convert", command=convert)
convert_button.pack()

# Output text area
output_text = tk.Text(root, height=10, width=50, state=tk.DISABLED)
output_text.pack(pady=10)

root.mainloop()
```
