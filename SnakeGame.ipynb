```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 640, 480
SNAKE_SIZE = 20
FPS = 10

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# Create the screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Snake Game")

# Snake initial position
snake_x, snake_y = WIDTH // 2, HEIGHT // 2
snake_speed_x, snake_speed_y = 0, 0

# Snake body
snake_body = [(snake_x, snake_y)]

# Food position
food_x, food_y = random.randint(0, (WIDTH - SNAKE_SIZE) // SNAKE_SIZE) * SNAKE_SIZE, random.randint(0, (HEIGHT - SNAKE_SIZE) // SNAKE_SIZE) * SNAKE_SIZE

# Score
score = 0

# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and snake_speed_y == 0:
                snake_speed_x, snake_speed_y = 0, -SNAKE_SIZE
            if event.key == pygame.K_DOWN and snake_speed_y == 0:
                snake_speed_x, snake_speed_y = 0, SNAKE_SIZE
            if event.key == pygame.K_LEFT and snake_speed_x == 0:
                snake_speed_x, snake_speed_y = -SNAKE_SIZE, 0
            if event.key == pygame.K_RIGHT and snake_speed_x == 0:
                snake_speed_x, snake_speed_y = SNAKE_SIZE, 0

    # Move the snake
    snake_x += snake_speed_x
    snake_y += snake_speed_y

    # Check for collisions with food
    if snake_x == food_x and snake_y == food_y:
        score += 1
        food_x, food_y = random.randint(0, (WIDTH - SNAKE_SIZE) // SNAKE_SIZE) * SNAKE_SIZE, random.randint(0, (HEIGHT - SNAKE_SIZE) // SNAKE_SIZE) * SNAKE_SIZE
    else:
        snake_body.pop()

    snake_body.insert(0, (snake_x, snake_y))

    # Check for collisions with walls or itself
    if snake_x < 0 or snake_x >= WIDTH or snake_y < 0 or snake_y >= HEIGHT or (snake_x, snake_y) in snake_body[1:]:
        running = False

    # Clear the screen
    screen.fill(BLACK)

    # Draw the snake
    for segment in snake_body:
        pygame.draw.rect(screen, WHITE, [segment[0], segment[1], SNAKE_SIZE, SNAKE_SIZE])

    # Draw the food
    pygame.draw.rect(screen, WHITE, [food_x, food_y, SNAKE_SIZE, SNAKE_SIZE])

    # Update the display
    pygame.display.update()

    # Control the game speed
    pygame.time.Clock().tick(FPS)

# Game over message
font = pygame.font.Font(None, 36)
game_over_text = font.render("Game Over", True, WHITE)
game_over_rect = game_over_text.get_rect(center=(WIDTH // 2, HEIGHT // 2))
screen.blit(game_over_text, game_over_rect)
pygame.display.update()

# Delay for a few seconds before exiting
pygame.time.delay(2000)

# Quit Pygame
pygame.quit()
```
